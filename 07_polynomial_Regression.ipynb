# polynomial Regression

```python
# bad fit
import numpy 
import matplotlib.pyplot as plt
x=[87,95,77,33,24,78,35,89,54,34,78,44,7,4,34]
y=[23,54,6,37,6,34,56,35,34,67,33,22,98,78,55,]
mymodel=numpy.poly1d(numpy.polyfit(x,y,3))
myline=numpy.linspace(2,95,100)
plt.scatter(x,y)
plt.plot(myline,mymodel(myline))
plt.show()
```

```python
import numpy
from sklearn.metrics import r2_score
x=[23,54,60,37,60,34,56,35,34,67,33,22,98,78,55]
y=[87,50,77,33,24,78,35,89,54,34,78,44,70,40,45]
model=numpy.poly1d(numpy.polyfit(x ,y, 3))
print(r2_score(y,model(x)))
```
